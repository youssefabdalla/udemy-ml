# K-Means Clustering

## Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
```

## Importing the dataset

```python
customers_data = pd.read_csv("Mall_Customers.csv")
X = customers_data.iloc[:, [3, 4]].to_numpy()
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans

wcss = []
for clusters_number in range(1, 11):
    kmeans = KMeans(n_clusters= clusters_number, init= 'k-means++', random_state= 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow method')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()
```

## Training the K-Means model on the dataset

```python
kmeans_specific_clusters = KMeans(n_clusters= 5, init= 'k-means++', random_state= 42)
y_k_means = kmeans_specific_clusters.fit_predict(X)
```

## Visualising the clusters

```python
plt.scatter(
    X[y_k_means == 0, 0], X[y_k_means == 0, 1], s=100, c="red", label="cluster_0"
)  # cluster 0

plt.scatter(
    X[y_k_means == 1, 0], X[y_k_means == 1, 1], s=100, c="green", label="cluster_1"
)  # cluster 1

plt.scatter(
    X[y_k_means == 2, 0], X[y_k_means == 2, 1], s=100, c="blue", label="cluster_2"
)  # cluster 2

plt.scatter(
    X[y_k_means == 3, 0], X[y_k_means == 3, 1], s=100, c="cyan", label="cluster_3"
)  # cluster 3

plt.scatter(
    X[y_k_means == 4, 0], X[y_k_means == 4, 1], s=100, c="magenta", label="cluster_4"
)  # cluster 4

plt.scatter(
    kmeans_specific_clusters.cluster_centers_[:, 0],
    kmeans_specific_clusters.cluster_centers_[:, 1],
    s=300,
    c="yellow",
    label="centroid",
)
plt.title('clusters of customers')
plt.xlabel('annual income (k$)')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
