```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Position_Salaries.csv")
X = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

Simple linear regression model

```python
from sklearn.linear_model import LinearRegression

linearRegressor = LinearRegression()
linearRegressor.fit(X, y)
```

```python
y_linear_pred = linearRegressor.predict(X)
```

Polynomial Linear regression model

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=5)
X_poly = poly_reg.fit_transform(X)
linearRegressor_2 = LinearRegression()
linearRegressor_2.fit(X_poly, y)
y_linear_pred_2 = linearRegressor_2.predict(X_poly)
```

Visualize linear results

```python
plt.scatter(X, y, color="red")
plt.plot(X, y_linear_pred, c="blue")
plt.title('Truth or Bluff (Linear regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

Vsualize polynomial results

```python
plt.scatter(X, y, color="red")
plt.plot(X, y_linear_pred_2, c="blue")
plt.title('Truth or Bluff (Polynomial Linear regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
