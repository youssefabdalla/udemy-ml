# KNN classifier

## Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## Importing the dataset

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1].to_frame()
```

## Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, shuffle=True, test_size=0.25, random_state=0
)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the KNN model on the Training set

```python
from sklearn.neighbors import KNeighborsClassifier

clr = KNeighborsClassifier(p=2)
clr.fit(X_train, y_train)
```

## Predicting a new result

```python
X_predict = sc.transform([[32, 150000]])
y_predict = clr.predict(X_predict)
print(y_predict)
```

## Predicting the Test set results

```python
y_test_predict = clr.predict(X_test)
print(
    np.concatenate(
        (
            y_test_predict.reshape(len(y_test_predict), 1),
            y_test.to_numpy().reshape(len(y_test), 1),
        ),
        axis=1,
    )
)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_test_predict, normalize='all'))
print(accuracy_score(y_test, y_test_predict, normalize=True))
```
