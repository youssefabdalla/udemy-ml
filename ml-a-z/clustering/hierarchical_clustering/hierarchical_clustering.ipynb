# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
# same would be from scipy.cluster import hierarchy
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset with 5 clusers

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

## Visualising the choice of 5 clusters

```python
plt.scatter(
    X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c="red", label="cluster_0"
)  # cluster 0

plt.scatter(
    X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c="green", label="cluster_1"
)  # cluster 1

plt.scatter(
    X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c="blue", label="cluster_2"
)  # cluster 2

plt.scatter(
    X[y_hc == 3, 0], X[y_hc == 3, 1], s=100, c="cyan", label="cluster_3"
)  # cluster 3

plt.scatter(
    X[y_hc == 4, 0], X[y_hc == 4, 1], s=100, c="magenta", label="cluster_4"
)  # cluster 4

plt.title('5 clusters of customers')
plt.xlabel('annual income (k$)')
plt.ylabel('spending score')
plt.legend()
plt.show()
```

## Training the Hierarchical Clustering model on the dataset with 3 clusers

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

## Visualising the choice of 3 clusters

```python
plt.scatter(
    X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c="red", label="cluster_0"
)  # cluster 0

plt.scatter(
    X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c="green", label="cluster_1"
)  # cluster 1

plt.scatter(
    X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c="blue", label="cluster_2"
)  # cluster 2

plt.title('3 clusters of customers')
plt.xlabel('annual income (k$)')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
